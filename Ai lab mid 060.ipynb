```python
fullcosts={}
def kharchapani(actualcookingtimeinmin=120,timeinminutes=210,rateoflpg=325,petrolcost=6000):
    
    l=["km","tv","tb","jb","kl"]
    
    heights={
        'km':7800,
        'tv':8100,
        'tb':9050,
        'jb':10170,
        'kl':11500
    }
    efs={}
    for i in l:
        if heights[i] in range(6000,7000):
            efs[i]=efs.get(i,1.20)
        elif heights[i] in range(7000,8000):
            efs[i]=efs.get(i,1.25)
        elif heights[i] in range(8000,9000):
            efs[i]=efs.get(i,1.30)
        elif heights[i] in range(9000,10000):
            efs[i]=efs.get(i,1.35)
        elif heights[i] in range(10000,11000):
            efs[i]=efs.get(i,1.40)
        elif heights[i] in range(11000,12000):
            efs[i]=efs.get(i,1.45)
        else:
            efs[i]=efs.get(i,0)
    
    print("ElevationFactors",efs)
    
    tourhighaltitudecookingtime={}
    for i in l:
        tourhighaltitudecookingtime[i]=tourhighaltitudecookingtime.get(i,(efs[i]*actualcookingtimeinmin))
    print("HighAltitudeCookingTime",tourhighaltitudecookingtime)
    
    

    
    
    lpgcosts={}
    lpgUsedInkg={}
    for i in l:
        lpgcosts[i]=lpgcosts.get(i,(rateoflpg*tourhighaltitudecookingtime[i]/timeinminutes))
        lpgUsedInkg[i]=lpgUsedInkg.get(i,(tourhighaltitudecookingtime[i]/timeinminutes))
    print("LpgCosts",lpgcosts)
    print("LpgUsedinKg",lpgUsedInkg)
    

    
    hotelcosts={
        'km':16000,
        'tv':12000,
        'tb':4000,
        'jb':4000,
        'kl':4000
    }
    
    
    
    for i in l:
        fullcosts[i]=fullcosts.get(i,(lpgcosts[i]+hotelcosts[i]+petrolcost))
    print("FullCosts",fullcosts)
    TotalExpense = sum(fullcosts.values())
    print("TotalExpense : ",TotalExpense)

    for i,(ct,a,c) in enumerate(list(zip(tourhighaltitudecookingtime.values(), lpgUsedInkg.values(),lpgcosts.values()))):
        print("Day "+ str(i+1) +" : " +str(ct) +" min %.2f" %a + " Kg %.2f" %c+" Rs")
    
    print("\nHotel charges")
    for i,ch in enumerate(hotelcosts.values()):
        print("Day "+ str(i+1) +" : Camp -- ", ch)
        
    return TotalExpense
    
Te=kharchapani()
```
